# Neural Networks

## Step 1. Importing Libraries

```python
import torch            #--> tensors
import torch.nn as nn        #--> neural network layers
import torch.optim as optim     #--> optimizer
```

## Step 2. Dataset

```python
# study hours and attendance
X = torch.tensor([
    [1.0, 30.0],
    [2.0, 40.0],
    [3.0, 50.0],
    [4.0, 60.0],
    [5.0, 70.0],
    [6.0, 80.0]
])
# pass/fail
y = torch.tensor([[0.0] , [0.0], [0.0] , [1.0] , [1.0] , [1.0]])
```

## Step 3. making class

```python
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel , self).__init__()
        # initializing values to be used in forward
        self.fc1 = nn.Linear(2,10)       # --> ye pehlay sikhata hai means k 2 input feautures hain or hum 10 diff calculation krein gay
        self.relu = nn.ReLU()           # --> ye model ko non linear banaye ga means k sirf straight line data nai bal k har type k liye suitable
        self.fc2 = nn.Linear(10,1)       #--> ye phir un 5 calculations mei se output dega valid
        self.sigmoid = nn.Sigmoid()     #--> ye un values ko binary mei convert kray ga jesay agar 0.8 ata hai to usay 1 , 0.3 ata hai to usay 0

    def forward(self , x):      #--> ye wala function sb above kam kray ga
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

## Step 4. Model , Loss & Optimizer

```python
model = MyModel()
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters() , lr=0.01)
```

## Step 5. Training Loop

```python
for epoch in range(1000):
    y_predicts = model(X)
    loss = loss_fn(y_predicts , y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f"Epoch:{epoch} , loss:{loss.item()}")
```

## Step 6. Prediction

```python
test = torch.tensor([[4.5 , 75.0]])
prediction = model(test)
print(f"Prediction:{prediction.item()}")
if prediction.item() >= 0.5:
    print("Pass ho ga")
else:
    print("Fail bruh!")
```

## Step 7. Saving Model

```python
torch.save(model.state_dict(), "Linear_Model.pth")
```
